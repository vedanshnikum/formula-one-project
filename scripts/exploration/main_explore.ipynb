```python
dbutils.widgets.text("schema", "default", "Schema")
```

```python
schema = dbutils.widgets.get("schema")
data = spark.catalog.listTables(f"formone.{schema}")
```

```python
%skip
for table in data:
    df = spark.table(f"formone.{schema}.{table.name}")
    print(f"{table.name}: {df.count()}")
    df.printSchema()
    display(df.limit(1))
```

```python
for i, table in enumerate(data):
    print(f"{i+1}: {table.name}")
```

```python
def inspect(data, file, index, show_table, show_schema, show_full):
    name = data[file].name

    print(f"Current File: {name}")
    print(f"Total Files: {len(data)}")
    print("="*20)
    df = spark.table(f"formone.{schema}.{name}")

    col = df.columns[index]
    print(f"Total Rows: {df.count()}")
    print(f"Total Columns: {len(df.columns)}")
    print(f"Data Type: {df.schema[index].dataType}")
    if(show_schema):
        df.printSchema()
    if (show_full):
        display(df)
    elif(show_table):
        display(df.limit(5))
    else:
        display(df.select(col).distinct().orderBy(col))
```

```python
show_table = False  
show_schema = True
show_full = False

file = 7
column = 5

inspect(data, file - 1, column - 1, show_table, show_schema, show_full)
```
