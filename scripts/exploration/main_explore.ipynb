#Flexible Exploration Notebook for Data in Any Layer

```python
schema = "gold"
data = spark.catalog.listTables(f"formone.{schema}")
```

```python
def inspect(data, file, index, show_table, show_schema, show_full, use_show):
    name = data[file].name

    print(f"Current File: {name}")
    print(f"Total Files: {len(data)}")
    print("="*20)
    df = spark.table(f"formone.{schema}.{name}")

    col = df.columns[index]
    print(f"Total Rows: {df.count()}")
    print(f"Total Columns: {len(df.columns)}")
    print(f"Data Type: {df.schema[index].dataType}")
    if(show_schema):
        df.printSchema()
    if (show_full):
        if (use_show):
            df.show()
        else:
            display(df)
    elif(show_table):
        if (use_show):
            df.show(5)
        else:
            display(df.limit(5))
    else:
        if (use_show):
            df.select(col).distinct().orderBy(col).show()
        else:
            display(df.select(col).distinct().orderBy(col))
```

```python
show_table = False  
show_schema = True
show_full = False
show_all = True
use_show = True

file = 8
column = 0

if show_all:
    for i, table in enumerate(data):
        inspect(data, i, 0, True, True, False, True)
else:
    inspect(data, file - 1, column - 1, show_table, show_schema, show_full, use_show)
```
