```python
schema = dbutils.widgets.get("schema")
data = spark.catalog.listTables(f"formone.{schema}")
passed_tables = []
failed_tables = []
```

```python
for table in data:
    df = spark.table(f"formone.{schema}.{table.name}")
    passed, failed_checks = validate_table(table.name, df)
    if passed:
        passed_tables.append(table.name)
    else:
        failed_tables.append(table.name)

_summary(passed_tables, failed_tables)
```
